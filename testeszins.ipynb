````python
#  ANALISADOR DE DESEMPENHO - Desafio IA
#  Curso de IA - CETAM / Amazonas
# =============================================================

TOTAL_MODELOS = 10   # número fixo de modelos

# --- Variáveis iniciais --------------------------------------
soma        = 0.0
maior       = -1.0   # começa baixo para qualquer valor substituir
menor       = 101.0  # começa alto para qualquer valor substituir

# --- Cabeçalho -----------------------------------------------
print("=" * 45)
print("    ANALISADOR DE DESEMPENHO - MODELOS IA")
print("=" * 45)
print("Digite a precisão de cada modelo (0 a 100).\n")

# --- Laço FOR: repete exatamente 10 vezes --------------------
for i in range(1, TOTAL_MODELOS + 1):

    while True:   # fica pedindo até o usuário digitar um valor válido
        entrada = input(f"  Precisão do Modelo {i:02d}: ")

        try:
            precisao = float(entrada)
        except ValueError:
            print("  ⚠  Digite apenas números. Tente novamente.\n")
            continue

        if precisao < 0 or precisao > 100:
            print("  ⚠  O valor deve estar entre 0 e 100.\n")
            continue

        break   # valor válido, sai do while interno

    # Acumula para calcular média, maior e menor
    soma += precisao

    if precisao > maior:
        maior = precisao

    if precisao < menor:
        menor = precisao

# --- Cálculo da média ----------------------------------------
media = soma / TOTAL_MODELOS

# --- Exibição dos resultados ---------------------------------
print("\n" + "=" * 45)
print("         RESULTADO DA ANÁLISE 📊")
print("=" * 45)
print(f"  Modelos avaliados  : {TOTAL_MODELOS}")
print(f"  Média das precisões: {media:.2f}%")
print(f"  Maior precisão     : {maior:.2f}%  ↑")
print(f"  Menor precisão     : {menor:.2f}%  ↓")
print("=" * 45)
````

## Laço de Repetição `while`

O laço `while` executa um bloco de código repetidamente **enquanto** uma condição especificada for verdadeira. A condição é verificada a cada iteração, e se ela se tornar falsa, o laço é encerrado.

### Estrutura básica:

```python
while condição:
    # bloco de código a ser executado enquanto a condição for verdadeira
    # É importante que o bloco de código altere algo que possa fazer a condição se tornar falsa em algum momento,
    # caso contrário, teremos um loop infinito.
```

````python
# Exemplo de laço while: Contagem regressiva

contador = 5

while contador > 0:
    print(f"Contagem regressiva: {contador}")
    contador -= 1  # Decrementa o contador para que a condição eventualmente se torne falsa

print("Fim da contagem!")
````

## Laço de Repetição `for`

O laço `for` é usado para iterar sobre uma sequência (como uma lista, tupla, string, ou um intervalo de números) ou outros objetos iteráveis. Ele executa um bloco de código para cada item da sequência.

### Estrutura básica:

```python
for item in sequencia:
    # bloco de código a ser executado para cada item
```

### Usando `range()` com `for`:

É comum usar `for` com a função `range()` para iterar um número específico de vezes. A função `range()` gera uma sequência de números.

- `range(stop)`: Gera números de 0 até `stop-1`.
- `range(start, stop)`: Gera números de `start` até `stop-1`.
- `range(start, stop, step)`: Gera números de `start` até `stop-1`, incrementando a cada `step`.

````python
# Exemplo de laço for: Iterando sobre uma lista
frutas = ["maçã", "banana", "cereja"]
for fruta in frutas:
    print(f"Eu gosto de {fruta}")

print("\n")

# Exemplo de laço for com range(): Contando de 1 a 5
print("Contagem de 1 a 5:")
for i in range(1, 6):
    print(i)

print("\n")

# Exemplo de laço for com range(): Contagem regressiva de 10 a 0, de 2 em 2
print("Contagem regressiva de 10 a 0 (de 2 em 2):")
for i in range(10, -1, -2):
    print(i)
````

## Exemplo de `for` para Acumular Valores e Calcular Média

Este exemplo demonstra como usar um laço `for` para pedir ao usuário um número fixo de valores, acumular a soma desses valores e, em seguida, calcular a média.

````python
9# Definimos quantos valores o usuário irá inserir
quantidade_valores = 3

soma_valores = 0.0

print(f"Por favor, digite {quantidade_valores} números para calcular a média.")

for i in range(1, quantidade_valores + 1):
    while True:
        try:
            valor = float(input(f"Digite o {i}º número: "))
            soma_valores += valor
            break # Sai do loop interno se o valor for válido
        except ValueError:
            print("Entrada inválida. Por favor, digite um número.")

# Calcula a média (evitando divisão por zero, embora neste caso a quantidade seja fixa)
if quantidade_valores > 0:
    media = soma_valores / quantidade_valores
    print(f"\nA soma dos valores é: {soma_valores:.2f}")
    print(f"A média dos valores é: {media:.2f}")
else:
    print("Nenhum valor foi inserido.")
````
