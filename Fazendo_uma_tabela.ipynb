```python
import pandas as pd

linhas_tabela = []

while True:
    nome = input("Digite o nome (ou 'sair' para finalizar a tabela): ")
    if nome.lower() == 'sair':
        break

    # Input validation for nota1
    while True:
        try:
            nota1 = float(input("Digite a nota 1 (entre 0 e 10): "))
            if 0 <= nota1 <= 10:
                break
            else:
                print("Nota 1 inválida. Por favor, digite um valor entre 0 e 10.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número para a nota 1.")

    # Input validation for nota2
    while True:
        try:
            nota2 = float(input("Digite a nota 2 (entre 0 e 10): "))
            if 0 <= nota2 <= 10:
                break
            else:
                print("Nota 2 inválida. Por favor, digite um valor entre 0 e 10.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número para a nota 2.")

    media = (nota1+nota2)/2

    resultado = ""
    if media >= 6:
            print("Aprovado")
            resultado = "Aprovado"
    else:
            print("Reprovado")
            resultado = "Reprovado"

    linhas_tabela.append({
        'Nome': nome,
        'nota1': float(nota1), # Nota 1 na tabela
        'nota2': float(nota2), # Nota 2 na tabela
        'media': float(media), # Media na tabela
        'Resultado': resultado, # Nova coluna para o resultado
    })

if linhas_tabela:
    df_usuario = pd.DataFrame(linhas_tabela)
    print("\nTabela com notas e media:")
    display(df_usuario)
else:
    print("Nenhum dado inserido para criar a tabela.")
```
