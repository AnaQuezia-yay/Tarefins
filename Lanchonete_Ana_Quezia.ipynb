Entradas:

1.   Quantidade de cada item do cardapio:

      *   Quantidade de Hamburguer/s

      *   Quantidade de Cheeseburguer/s

      *   Quantidade de Refrigerante/s

      *   Quantidade de Batatas Fritas


2.   Tempo de preparo de cada item do cardapio:

        *   Hamburguer e Cheeseburguer = 10 minutos

        *   Refrigerante = 2 minutos

        *   Batatas Fritas = 5 minutos


Saidas

1.   Total (sem desconto)

2.   Total com desconto

3.   Total a pagar

4.   Tempo de preparo








```python
# Solicita ao usuário a quantidade de cada opção do cardapio
Hamburguer = int(input("Quantidade de Hamburguer/s: "))
Cheeseburguer = int(input("Quantidade de Cheeseburguer/s: "))
Refrigerante = int(input("Quantidade de Refrigerante/s: "))
Batata_frita = int(input("Quantidade de Batata/s Frita/s: "))

Quant_Hamburguer = 15.00*Hamburguer
preparo_Hamburguer = 10*Quant_Hamburguer

Quant_Cheeseburguer = 18.00*Cheeseburguer
preparo_Cheeseburger = 10*Quant_Cheeseburguer

Quant_Refrigerante = 6.00*Refrigerante
preparo_Refrigerante = 2*Quant_Refrigerante

Quant_Batata_frita = 10.00*Batata_frita
preparo_Batata_frita = 5*Quant_Batata_frita


total = (Quant_Hamburguer + Quant_Cheeseburguer + Quant_Refrigerante + Quant_Batata_frita)
tempo = (preparo_Hamburguer + preparo_Cheeseburger + preparo_Refrigerante + preparo_Batata_frita)

desconto = 0

if total > 50:
    desconto = total*15/100

total_final = total - desconto


print("Total sem Desconto:", total, "Reais")
print("Desconto: R$", desconto)
print("Total a pagar:", total_final, "Reais")
print("Tempo de preparo:", tempo, "Minutos")
```
